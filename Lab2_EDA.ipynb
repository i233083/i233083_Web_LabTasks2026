```python
import pandas as pd
import seaborn as sns

# Load Titanic dataset
df_titanic = sns.load_dataset("titanic")

# Check missing values per column
df_titanic.isnull().sum()

# Mean imputation for age
df_titanic["age"] = df_titanic["age"].fillna(df_titanic["age"].mean())

# Mode imputation for embarked
df_titanic["embarked"] = df_titanic["embarked"].fillna(df_titanic["embarked"].mode()[0])

# Mode imputation for embark_town
df_titanic["embark_town"] = df_titanic["embark_town"].fillna(
    df_titanic["embark_town"].mode()[0]
)

# Drop deck column
df_titanic = df_titanic.drop(columns=["deck"])

# Verify missing values after preprocessing
df_titanic.isnull().sum()

df_titanic.info()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df_tips = sns.load_dataset("tips")

# Columns to check
columns = ["total_bill", "tip"]

for col in columns:
    Q1 = df_tips[col].quantile(0.25)
    Q3 = df_tips[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df_tips[(df_tips[col] < lower) | (df_tips[col] > upper)]
    print(f"{col.capitalize()} Outliers Count: {outliers.shape[0]}")

    # Box plot using Seaborn
    plt.figure()
    sns.boxplot(x=df_tips[col])
    plt.title(f"Box Plot of {col.capitalize()} (Tips Dataset)")
    plt.xlabel(col.capitalize())
    plt.show()
```

```python
# Task 8: Outlier Mitigation via IQR-Based Row Elimination

import seaborn as sns

# Load Tips dataset
df_tips = sns.load_dataset("tips")

# Dataset size BEFORE outlier removal
print("Dataset shape before outlier removal:", df_tips.shape)

# ---------- IQR bounds for total_bill ----------
Q1_total = df_tips['total_bill'].quantile(0.25)
Q3_total = df_tips['total_bill'].quantile(0.75)
IQR_total = Q3_total - Q1_total

lower_bound_total = Q1_total - 1.5 * IQR_total
upper_bound_total = Q3_total + 1.5 * IQR_total

# ---------- IQR bounds for tip ----------
Q1_tip = df_tips['tip'].quantile(0.25)
Q3_tip = df_tips['tip'].quantile(0.75)
IQR_tip = Q3_tip - Q1_tip

lower_bound_tip = Q1_tip - 1.5 * IQR_tip
upper_bound_tip = Q3_tip + 1.5 * IQR_tip

# ---------- Outlier Removal ----------
df_tips_cleaned = df_tips[
    (df_tips['total_bill'] >= lower_bound_total) &
    (df_tips['total_bill'] <= upper_bound_total) &
    (df_tips['tip'] >= lower_bound_tip) &
    (df_tips['tip'] <= upper_bound_tip)
]

# Dataset size AFTER outlier removal
print("Dataset shape after outlier removal:", df_tips_cleaned.shape)
```

```python
# Task 9: Post-Processing Statistical Reassessment

import pandas as pd
import seaborn as sns

# Load Tips dataset
df_tips = sns.load_dataset("tips")

# ---------- Statistics BEFORE Outlier Removal ----------
stats_before = pd.DataFrame({
    'Mean': df_tips[['total_bill', 'tip']].mean(),
    'Median': df_tips[['total_bill', 'tip']].median(),
    'Std_Dev': df_tips[['total_bill', 'tip']].std()
})

# ---------- IQR-based Outlier Removal (same as Task 8) ----------
Q1_total = df_tips['total_bill'].quantile(0.25)
Q3_total = df_tips['total_bill'].quantile(0.75)
IQR_total = Q3_total - Q1_total

lower_total = Q1_total - 1.5 * IQR_total
upper_total = Q3_total + 1.5 * IQR_total

Q1_tip = df_tips['tip'].quantile(0.25)
Q3_tip = df_tips['tip'].quantile(0.75)
IQR_tip = Q3_tip - Q1_tip

lower_tip = Q1_tip - 1.5 * IQR_tip
upper_tip = Q3_tip + 1.5 * IQR_tip

df_tips_cleaned = df_tips[
    (df_tips['total_bill'] >= lower_total) &
    (df_tips['total_bill'] <= upper_total) &
    (df_tips['tip'] >= lower_tip) &
    (df_tips['tip'] <= upper_tip)
]

# ---------- Statistics AFTER Outlier Removal ----------
stats_after = pd.DataFrame({
    'Mean': df_tips_cleaned[['total_bill', 'tip']].mean(),
    'Median': df_tips_cleaned[['total_bill', 'tip']].median(),
    'Std_Dev': df_tips_cleaned[['total_bill', 'tip']].std()
})

# ---------- Comparative Statistical Table ----------
comparative_stats = pd.concat(
    [stats_before, stats_after],
    axis=1,
    keys=['Before Outlier Removal', 'After Outlier Removal']
)

# Display comparative statistics
comparative_stats
```
