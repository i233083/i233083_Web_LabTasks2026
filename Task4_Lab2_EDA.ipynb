```python
import seaborn as sns
import pandas as pd
df_iris = sns.load_dataset("iris")
df_titanic = sns.load_dataset("titanic")
df_tips = sns.load_dataset("tips")
print("IRIS DATASET")

iris_duplicates = df_iris.duplicated().sum()
print("Number of duplicate records:", iris_duplicates)

print("Shape before duplicate removal:", df_iris.shape)

df_iris_cleaned = df_iris.drop_duplicates()

print("Shape after duplicate removal:", df_iris_cleaned.shape)
print("\nTITANIC DATASET")

titanic_duplicates = df_titanic.duplicated().sum()
print("Number of duplicate records:", titanic_duplicates)

print("Shape before duplicate removal:", df_titanic.shape)

df_titanic_cleaned = df_titanic.drop_duplicates()

print("Shape after duplicate removal:", df_titanic_cleaned.shape)
print("\nTIPS DATASET")

tips_duplicates = df_tips.duplicated().sum()
print("Number of duplicate records:", tips_duplicates)

print("Shape before duplicate removal:", df_tips.shape)

df_tips_cleaned = df_tips.drop_duplicates()

print("Shape after duplicate removal:", df_tips_cleaned.shape)

```
