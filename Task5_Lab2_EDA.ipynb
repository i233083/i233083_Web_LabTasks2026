```python
import seaborn as sns
import pandas as pd

df_iris = sns.load_dataset("iris")
df_titanic = sns.load_dataset("titanic")
df_tips = sns.load_dataset("tips")

# IRIS DATASET
print("IRIS DATASET: Species-wise Mean and Standard Deviation")
iris_stats = df_iris.groupby("species").agg(["mean", "std"])
print(iris_stats)

# TITANIC DATASET
print("\nTITANIC DATASET: Survival-wise Mean of Age and Fare")
titanic_stats = df_titanic.groupby("survived")[["age", "fare"]].mean()
print(titanic_stats)

# TIPS DATASET
print("\nTIPS DATASET: Mean and Median of Total Bill and Tip")
tips_stats = df_tips[["total_bill", "tip"]].agg(["mean", "median"])
print(tips_stats)
```
