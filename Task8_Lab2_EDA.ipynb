```python
# Task 8: Outlier Mitigation via IQR-Based Row Elimination

import seaborn as sns

# Load Tips dataset
df_tips = sns.load_dataset("tips")

# Dataset size BEFORE outlier removal
print("Dataset shape before outlier removal:", df_tips.shape)

# ---------- IQR bounds for total_bill ----------
Q1_total = df_tips['total_bill'].quantile(0.25)
Q3_total = df_tips['total_bill'].quantile(0.75)
IQR_total = Q3_total - Q1_total

lower_bound_total = Q1_total - 1.5 * IQR_total
upper_bound_total = Q3_total + 1.5 * IQR_total

# ---------- IQR bounds for tip ----------
Q1_tip = df_tips['tip'].quantile(0.25)
Q3_tip = df_tips['tip'].quantile(0.75)
IQR_tip = Q3_tip - Q1_tip

lower_bound_tip = Q1_tip - 1.5 * IQR_tip
upper_bound_tip = Q3_tip + 1.5 * IQR_tip

# ---------- Outlier Removal ----------
df_tips_cleaned = df_tips[
    (df_tips['total_bill'] >= lower_bound_total) &
    (df_tips['total_bill'] <= upper_bound_total) &
    (df_tips['tip'] >= lower_bound_tip) &
    (df_tips['tip'] <= upper_bound_tip)
]

# Dataset size AFTER outlier removal
print("Dataset shape after outlier removal:", df_tips_cleaned.shape)
```
